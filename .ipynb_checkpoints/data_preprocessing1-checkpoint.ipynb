### Importing the Libraries

```python
import pandas as pd
import numpy as np
```

### Importing the Dataset

```python
movies_df = pd.read_csv("Dataset/tmdb_5000_movies.csv")
```

```python
movies_df.head()
```

```python
credits_df = pd.read_csv("Dataset/tmdb_5000_credits.csv")
```

```python
credits_df.head()
```

```python
# Renaming the 'id' column in the movies dataframe with 'movie_id'
movies_df.rename(columns= {'id': 'movie_id'}, inplace= True)
```

```python
#movies_df.head()
```

```python
# Merging the two dataset
movies_df = movies_df.merge(credits_df, on= 'movie_id')
```

```python
movies_df.head()
```

```python
movies_df.columns
```

```python
# Keeping the relevant columns
movies_df = movies_df.drop(columns= ['budget', 'homepage', 'keywords', 'original_language',
                                     'popularity', 'production_companies', 'production_countries',
                                     'release_date', 'revenue', 'runtime', 'spoken_languages', 'status',
                                     'tagline', 'title_x', 'title_y', 'vote_average', 'vote_count', 'crew',])
```

```python
movies_df.head()
```

```python
movies_df.iloc[0,0]
```

```python
# Converting the string of brackets and curly braces into lists of dictionary objects
from ast import literal_eval
elements = ['genres', 'cast']
for element in elements:
    movies_df[element] = movies_df[element].apply(literal_eval)
```

```python
movies_df.loc[0, 'cast']
```

```python
# Taking out the cast and genres names from the lists
def get_name(data):
    if isinstance(data, list):
        names = [ col['name'] for col in data ]
        if len(names) > 3:
            names = names[:3]
        return names
    return []  # returning empty list in case of missing data
```

```python
for element in elements:
    movies_df[element] = movies_df[element].apply(get_name)
```

```python
movies_df.head()
```

```python
movies_df.shape
```

```python
# Checking the null values
movies_df.isnull().sum()
```

```python
movies_df = movies_df.dropna()
```
